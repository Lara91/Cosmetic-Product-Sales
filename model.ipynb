## 4.Explanatory Data Analysis:


**Product Category_Sales Performance/Category Revenue Comparison**

4.1 Which product categories bring in the most money all around?

By identifying the product categories that provide the *highest sales*, my company can efficiently assign resources and focus its marketing efforts on these categories .

```python
df.head()
```

```python
# grouping the data for net sales for every Category Name ID
category_revenue = df.groupby('Category Name ID')['Net Sales calculated'].sum().reset_index()

# Sorting the data by 'Net Sales calculated'
category_revenue = category_revenue.sort_values('Net Sales calculated', ascending=False)

# Plotting  the data
plt.bar(category_revenue['Category Name ID'].astype(str), category_revenue['Net Sales calculated'])
plt.title('Revenue by Category')
plt.xlabel('Category Name ID')
plt.ylabel('Net Sales')
plt.show()
```

**Analyzing Delhi(DL) areas Sales**

4.2 Do the sales volumes in the different t zones in  Delhi differ significantly from one other?

Analyzing zones sales patterns helps  us de loping targeted marketing strategies specific to the needs and preferences of customers in each area.

 


```python
# i will  group the data by "Zone " .
# I will calculate the total net sales for each zone.
# The results are sorted based on sales (highest to lowest).

regional_sales = df.groupby("Zone")["Net Sales calculated"].sum().reset_index()
regional_sales.sort_values(by="Net Sales calculated", ascending=False, inplace=True)

# Plotting sales by Zone
plt.figure(figsize=(10, 6))
plt.bar(regional_sales['Zone'], regional_sales['Net Sales calculated'])
plt.title('Sales Distribution by Zone')
plt.xlabel('Zone')
plt.ylabel('Net Sales Calculated')
plt.xticks(rotation=90)
plt.show()
```

 **Monthly Sales Trends:**
 
 4.3.How do monthly net sales change throughout the year?

Showing which months have the low sales help us to improve our markting statigrs to increas the sales in those months :

```python
# Grouping the sales by month 
monthly_sales = df.groupby('Month')['Net Sales calculated'].sum().reset_index()


# sorting by month January -> December
monthly_sales = monthly_sales.sort_values('Month')

# Plotting using matplot
plt.plot(monthly_sales['Month'], monthly_sales['Net Sales calculated'], marker='o')
plt.title('Monthly Net Sales Trends')
plt.xlabel('Month')
plt.ylabel('Net Sales')
plt.xticks(range(1, 13), ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.show()
```

**Discount Impact on Sales Volume**

4.4 Is there a noticeable trend between the discount provided and the quantity sold? 

This helps us to see if the discount is benefical for the sales .

```python
correlation = df['Cash Discount'].corr(df['Qty'])
print(f"Correlation coefficient between Cash Discount and Quantity Sold: {correlation}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cash Discount', y='Qty', data=df)
plt.title('Relationship between Cash Discount and Quantity Sold')
plt.xlabel('Cash Discount')
plt.ylabel('Quantity Sold')
plt.show()
```

As shown in the plot above , the discount does not effect the quanity 

**Top 5 Best-Selling Products**

4.5 What are the top 5 products in terms of quantity sold?


```python
# Group by 'ParentSKU' and sum 'Qty'
product_sales = df.groupby('ParentSKU')['Qty'].sum().reset_index()

# Sort the data by 'Qty' in descending order and select the top 5
top_products = product_sales.sort_values('Qty', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_products['ParentSKU'], top_products['Qty'])
plt.title('Top 5 Best-Selling Products by Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Price vs. Quantity Sold Analysis**

4.6 How does the price of a product relate to its sales volume?

Help us to know which products are not bring sold due to high price.

```python
# Creating a scatter plot to visualize the relationship

sns.scatterplot(x='Price', y='Qty', data=df)
plt.title('Price vs Quantity Sold')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.show()

correlation = df['Price'].corr(df['Qty'])
print("Correlation coefficient between Price and Quantity Sold: ", correlation)
```
